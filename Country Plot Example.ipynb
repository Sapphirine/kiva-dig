```python
from __future__ import unicode_literals
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.io.shapereader as shpreader
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

shapename = 'admin_0_countries'
countries_shp = shpreader.natural_earth(resolution='110m',
                                        category='cultural', name=shapename)

# some nice "earthy" colors
earth_colors = np.array([(199, 233, 192),
                                (161, 217, 155),
                                (116, 196, 118),
                                (65, 171, 93),
                                (35, 139, 69),
                                ]) / 255.

earth_colors = [
    cm.get_cmap('jet')(i) for i in range(0, 256, 8)
]
earth_colors = itertools.cycle(earth_colors)


ax = plt.axes(projection=ccrs.PlateCarree())
for country in shpreader.Reader(countries_shp).records():
    print(country.attributes['name'], earth_colors.next())
    ax.add_geometries(country.geometry, ccrs.PlateCarree(),
                      facecolor=earth_colors.next(),
                      label=unicode(country.attributes['sovereignt']))
plt.gcf().set_size_inches(18, 6)
```
